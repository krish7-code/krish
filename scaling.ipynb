```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MaxAbsScaler
from sklearn.preprocessing import RobustScaler

#load the dataset
df=pd.read_csv("winequality-red.csv")

#select the numeric column only
df=df.select_dtypes(include='number')
df.head()
```

```python
#handle the null values and duplicate values
c_df=df.drop_duplicates()
c_df=df.fillna(df.mean())
```

```python
# normalization ==> minmax scaler

scaler=MinMaxScaler()
scaled_df=scaler.fit_transform(c_df)
scaled_df=pd.DataFrame(scaled_df,columns=c_df.columns)
scaled_df.head()
```

```python
#Standardization
scaler1=StandardScaler()
scaled_df1=scaler1.fit_transform(c_df)
scaled_df1=pd.DataFrame(scaled_df1,columns=c_df.columns)
scaled_df1.head()
```

```python
# Robust Scaling
scaler = RobustScaler()
scaled_df3 = scaler.fit_transform(c_df)
scaled_df3 = pd.DataFrame(scaled_df3, columns=c_df.columns)
scaled_df3.head()
```

```python
#MaxAbs Scaling
scaler= MaxAbsScaler()
scaled_df2 = scaler.fit_transform(c_df)
scaled_df2 = pd.DataFrame(scaled_df2, columns=c_df.columns)
scaled_df2.head()
```
